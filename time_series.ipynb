```python
import pandas as pd, matplotlib.pyplot as plt, coronavirus
from coronavirus import getData, getPlot
%matplotlib inline
```

```python
asia = [
    'Thailand', 'Singapore', 'Taiwan*', 'Hong Kong', 'Macau', 
    'Malaysia', 'Pakistan', 'Vietnam', 'Indonesia', 'Australia', 
    'China', 'Korea, South', 'Japan', 'Philippines', 'India', 
]
europe = [
    'Italy', 'Spain', 'Germany', 'France', 
    'United Kingdom', 'Switzerland', 'Belgium', 'Netherlands', 
]
america = ['US', 'Canada', 'Brazil', 'Peru', 'Ecuador', ]
middleEast = ['Iran', 'Turkey', 'Israel', 'Saudi Arabia', 'United Arab Emirates', ]
used = asia + europe + america + middleEast
```

```python
source = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
world = source.drop(['Province/State', 'Lat', 'Long'], axis = 1).groupby('Country/Region').sum()
other = source[source['Province/State'].isin(used)].drop(['Country/Region', 'Lat', 'Long'], axis=1).groupby('Province/State').sum()
coronavirus.both1 = pd.concat([world, other])
```

```python
coronavirus.gap = 7
coronavirus.writer = pd.ExcelWriter('time_series.xlsx')
coronavirus.both2 = getData(coronavirus.both1, 'world')
coronavirus.writer.save()
```
